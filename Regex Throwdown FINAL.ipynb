```python
align1=AlignIO.read("C:\\Users\\BioInf_AO\\Downloads\\user1_aligned.fasta","fasta")
align2=AlignIO.read ("C:\\Users\\BioInf_AO\\Downloads\\user2_aligned.fasta", "fasta")
alignselection=AlignIO.read("C:\\Users\\BioInf_AO\\Downloads\\SLC26A1alignmentfive.fasta", "fasta")

print("Here are the alignments:\n\n")
print(User1,"'s alignment:")
print(align1)
print("\n\n")
print(User2,"'s alignment:")
print (align2)
print("\n\n")
print("Here is the alignment generated from the standard consensus pattern:")
print (alignselection)
```

```python
print("Here are the distance matrices generated:")

print(User1,"'s matrix:")
calculator=DistanceCalculator('identity')
Main1=calculator.get_distance(align1)
Main1
print(Main1)

print("\n\n")

print(User2,"'s matrix:")
calculator=DistanceCalculator('identity')
Main2=calculator.get_distance(align2)
Main2
print(Main2)

print("\n\n")

print("The distance matrix generated from the selected consensus pattern:")
calculator=DistanceCalculator('identity')
Mainselection=calculator.get_distance(alignselection)
Mainselection
print(Mainselection)
```
